```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def generate_se_tasks_chart():
    df = pd.read_csv('tasks.csv')
    plt.figure(figsize=(14, 8))
    plt.rcParams.update({'font.size': 18})
    df_sorted = df.sort_values(by='Numero de Referencias', ascending=True)
    bars = plt.barh(df_sorted['Tarefa de ES'], df_sorted['Numero de Referencias'])
    
    # plt.title('Tarefas de Eng. de Software', fontsize=14)
    plt.xlabel('Num. de Referencias')
    plt.ylabel('Tarefa')
    
    for bar in bars:
        width = bar.get_width()
        plt.text(width + (width * 0.02),
                 bar.get_y() + bar.get_height()/2.,
                 f'{int(width)}', 
                 ha='left', va='center')

    plt.locator_params(axis='x', nbins=5) 
    plt.savefig('tarefas_ranking.png', dpi=300, bbox_inches='tight', transparent=False)
    plt.show()

def generate_publication_date_chart():
    """Generate horizontal bar chart for publication dates"""
    df = pd.read_csv('papers.csv')
    plt.figure(figsize=(16, 8))
    plt.rcParams.update({'font.size': 16})
    publication_counts = df['Publication Date'].value_counts()
    publication_counts_sorted = publication_counts.sort_values(ascending=True)
    bars = plt.barh(publication_counts_sorted.index, publication_counts_sorted.values)
    
    # plt.title('Estudos por data de publicacao', fontsize=14)
    plt.xlabel('Num. de Estudos')
    plt.ylabel('Data de Publicacao')
    
    for bar in bars:
        width = bar.get_width()
        plt.text(width + (width * 0.02),
                 bar.get_y() + bar.get_height()/2.,
                 f'{int(width)}', 
                 ha='left', va='center')
    
    plt.savefig('publication_dates.png', dpi=300, bbox_inches='tight', transparent=False)
    plt.show()

generate_se_tasks_chart()
generate_publication_date_chart()
```
